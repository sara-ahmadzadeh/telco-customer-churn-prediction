```python
# --- Data Cleaning & Feature Engineering ---
# Import the necessary library
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Load the data
df = pd.read_csv('../data/raw/telco_churn.csv')  # The path to raw dataset

# 1. Drop the customer ID (not useful for prediction)
df = df.drop('customerID', axis=1)
print("Dropped customerID. Shape:", df.shape)

# 2. Convert TotalCharges to numeric (we already did this)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())

# 3. Separate features (X) and target (y)
X = df.drop('Churn', axis=1)  # Features
y = df['Churn']               # Target

# 4. Identify categorical columns
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()
print(f"Categorical columns to encode: {categorical_cols}")

# 5. One-hot encode using sklearn (recommended for professional projects)
encoder = OneHotEncoder(drop='first', sparse_output=False)
X_encoded = encoder.fit_transform(X[categorical_cols])

# Get column names for the encoded data
encoded_columns = encoder.get_feature_names_out(categorical_cols)
X_encoded_df = pd.DataFrame(X_encoded, columns=encoded_columns, index=X.index)

# 6. Combine with numerical columns
X_final = pd.concat([X.drop(columns=categorical_cols), X_encoded_df], axis=1)

# 7. Convert target to binary (Yes=1, No=0)
y_binary = y.map({'Yes': 1, 'No': 0})

# 8. Check the final shapes
print(f"Final features shape: {X_final.shape}")
print(f"Final target shape: {y_binary.shape}")
print(f"Target distribution (1 = Churn):")
print(y_binary.value_counts())

# 9. Save the processed data for later use
X_final.to_csv('../data/processed/X_features.csv', index=False)
y_binary.to_csv('../data/processed/y_target.csv', index=False)
print("Processed data saved to data/processed/")
```
